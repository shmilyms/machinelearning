```python
X_train = np.array([[1,2],[2,3],[4,5],[3,1],[2,2],[12,12],[11,14],[13,13]])
y_train = np.array([0,0,0,0,0,1,1,1])
X_val  = np.array([[1,1],[1,4],[2,5],[11,11],[12,11]])
```

```python
class LogisticRegression:
    def __init__(self,learning_rate=0.1,n_iterations=1000):
        self.learning_rate=learning_rate
        self.n_iterations=n_iterations
    def __str__(self):
        return 'LogisticRegression(learning_rate={}, n_iterations={})'.format(self.learning_rate,self.n_iterations)
    def fit(self,X,y):
        m_samples,n_features=X.shape
        w=np.zeros(n_features+1).reshape(-1,1)
        X=np.insert(X,0,1,axis=1)
        y=y.reshape(-1,1)
        for _ in range(self.n_iterations):
            w-=self.learning_rate*gradient(w,X,y)
        self.w=w
        print(self)
    def predict(self,X):
        X=np.insert(X,0,1,axis=1)
        return sigmoid(self.w,X).flatten()
def sigmoid(w,X):
    return 1/(1+np.exp(-X.dot(w)))
def gradient(w,X,y):
    return X.T.dot(sigmoid(w,X)-y)/y.size
```

```python
plt.figure()
for n,it in enumerate([1,2,3,4,5,10,20,50,100,200,500,1000,10000]):
    reg=LogisticRegression(learning_rate=0.1,n_iterations=it)
    reg.fit(X_train,y_train);
    res=reg.predict(X_val)
    plt.plot(range(1,6),res,c=(0,0.5,1,((n/12)**1.8)*0.9+0.1))
```
